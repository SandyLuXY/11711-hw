```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
import pandas as pd
import seaborn as sns
```

```python
model_num = '150'
model_name = 'scibert_uncase'
# preds_tmp= np.array(np.load('outputs/'+model_name+'/test_labels_all_'+model_num+'.npy',allow_pickle=True))
# labs_tmp= np.array(np.load('outputs/'+model_name+'/test_labels_'+model_num+'.npy',allow_pickle=True))
preds_tmp= np.array(np.load('outputs/'+model_name+'/predict_'+model_num+'.npy',allow_pickle=True))
labs_tmp= np.array(np.load('outputs/'+model_name+'/labels_'+model_num+'.npy',allow_pickle=True))
cm1 = metrics.confusion_matrix(labs_tmp, preds_tmp,labels=range(0,8))

plt.figure(figsize=(9,9))
sns.heatmap(cm1, annot=True, fmt=".0f", linewidths=.5, square=True, cmap='Blues_r')
plt.ylabel('Label', size=17)
plt.xlabel('predicted number', size=17)
plt.title("%s acc: %.2f"%('epoch '+model_num,np.sum(preds_tmp==labs_tmp)*100.0/len(preds_tmp)))
plt.ylim(8+1,0)
plt.show()
```
